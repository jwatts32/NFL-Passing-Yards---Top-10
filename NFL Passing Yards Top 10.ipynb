```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
url_pass = "https://www.pro-football-reference.com/years/2023/passing_advanced.htm"
pass_stats = pd.read_html(url_pass, header=1)[0]
display(pass_stats.head(10))
```

```python
pass_stats = pass_stats[["Player", "Yds", "CAY/Cmp"]].head(10)
pass_stats = pass_stats.rename(columns={"CAY/Cmp":"AY/Cmp"})
display(pass_stats.sort_values("Yds").head(10))
```

```python
ax = pass_stats.plot(kind="scatter", x="AY/Cmp", y="Yds", fontsize=25, figsize=(30,15))
plt.title("2023 Yds & Air Yds Per Cmp", fontsize=50, color="blue")
ax.set_xlabel('AY/Cmp', fontsize = 40, color="blue") 
ax.set_ylabel('Yds', fontsize = 40, color="blue")

for index, row in pass_stats.iterrows():
    ax.annotate(row['Player'], (row['AY/Cmp'], row['Yds']), fontsize=30)    
```
